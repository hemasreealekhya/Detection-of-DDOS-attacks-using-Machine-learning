```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix
```

```python
# Step 1: Data Loading and Preprocessing
# Load the dataset
df = pd.read_csv("C:DDos.csv")
```

```python
df.head(3)
```

```python
# Remove the spaces before the column names
df.columns = df.columns.str.strip()
```

```python
# Drop any rows with missing values
df.dropna(inplace=True)

# Remove spaces before column names
df.columns = df.columns.str.strip()
df
```

```python
#Unique values in the Label target column

df.loc[:,'Label'].unique()
```

```python
#Checking the null values in the dataset.
plt.figure(1,figsize=( 10,4))
plt.hist( df.isna().sum())
# Set the title and axis labels
plt.xticks([0, 1], labels=['Not Null=0', 'Null=1'])
plt.title('Columns with Null Values')
plt.xlabel('Feature')
plt.ylabel('The number of features')

# Show the plot
plt.show()
```

```python
def plotMissingValues(dataframe):
    missing_values = dataframe.isnull().sum()  # Counting null values for each column
    fig = plt.figure(figsize=(16, 5))
    missing_values.plot(kind='bar')
    plt.xlabel("Features")
    plt.ylabel("Missing values")
    plt.title("Total number of Missing values in each feature")
    plt.show()

plotMissingValues(df)
```

```python
## Removing the null values
data_f=df.dropna()
```

```python
plt.figure(figsize=(10, 4))  # Adjusting the figure size
plt.hist(data_f.isna().sum())  # Plotting the histogram
plt.title('Data after removing the Null Values')  # Setting the title
plt.xlabel('The number of null values')  # Setting the x-axis label
plt.ylabel('Number of columns')  # Setting the y-axis label
plt.show()  # Displaying the plot
```

```python
pd.set_option('use_inf_as_na', True)  # Treat inf as NaN
null_values=data_f.isnull().sum()  # Check for NaN values
```

```python
# Convert the labels in the DataFrame to numerical values
data_f['Label'] = data_f['Label'].map({'BENIGN': 0, 'DDoS': 1})
```

```python
# Print the DataFrame

plt.hist(data_f['Label'], bins=[0, 0.3,0.7,1], edgecolor='black')  # Specify bins as [0, 1]
plt.xticks([0, 1], labels=['BENIGN=0', 'DDoS=1'])
plt.xlabel("Classes")
plt.ylabel("Count")
plt.show()
```

```python
# Convert into numpy array

#X1=np.array(data_f).astype(np.float64)
#y1=np.array(data_f['Label'])
# Split data into features and target variable
X = df.drop('Label', axis=1)
y = df['Label']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
print("The train dataset size = ",X_train.shape)
print("The test dataset size = ",X_test.shape)
```

```python
# Step 2: Model Training and Evaluation
# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=50, random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
# Getting feature importances from the trained model
importances = rf_model.feature_importances_

# Getting the indices of features sorted by importance
indices = sorted(range(len(importances)), key=lambda i: importances[i], reverse=False)
feature_names = [f"Features {i}" for i in indices]  # Replace with your column names

# Plotting feature importances horizontally
plt.figure(figsize=(8, 14))
plt.barh(range(X_train.shape[1]), importances[indices], align="center")
plt.yticks(range(X_train.shape[1]), feature_names)
plt.xlabel("Importance")
plt.title("Feature Importances")
plt.show()
```

```python
from sklearn.tree import plot_tree

estimator = rf_model.estimators_[0]  # Selecting the first estimator from the random forest model


plt.figure(figsize=(40, 20))
plot_tree(estimator, filled=True, rounded=True)
plt.show()
```

```python
# Evaluate Random Forest model
rf_accuracy = accuracy_score(y_test, rf_pred)
rf_f1 = f1_score(y_test, rf_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
rf_precision = precision_score(y_test, rf_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
rf_recall = recall_score(y_test, rf_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'

print('\nRandom Forest Metrics:')
print(f'Accuracy: {rf_accuracy:.4f}')
print(f'F1 Score: {rf_f1:.4f}')
print(f'Precision: {rf_precision:.4f}')
print(f'Recall: {rf_recall:.4f}')
```

```python
# Train Logistic Regression model
lr_model = LogisticRegression(random_state=42)
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

```python
# Evaluate Logistic Regression model
lr_accuracy = accuracy_score(y_test, lr_pred)
lr_f1 = f1_score(y_test, lr_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
lr_precision = precision_score(y_test, lr_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
lr_recall = recall_score(y_test, lr_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'

print('\nLogistic Regression Metrics:')
print(f'Accuracy: {lr_accuracy:.4f}')
print(f'F1 Score: {lr_f1:.4f}')
print(f'Precision: {lr_precision:.4f}')
print(f'Recall: {lr_recall:.4f}')
```

```python
# Train Neural Network model
nn_model = MLPClassifier(hidden_layer_sizes=(10,), max_iter=10, random_state=42)
nn_model.fit(X_train, y_train)

# Make predictions
nn_pred = nn_model.predict(X_test)

# Evaluate Neural Network model
nn_accuracy = accuracy_score(y_test, nn_pred)
nn_f1 = f1_score(y_test, nn_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
nn_precision = precision_score(y_test, nn_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
nn_recall = recall_score(y_test, nn_pred, pos_label='DDoS')  # Specify positive label as 'DDoS'
nn_confusion_matrix = confusion_matrix(y_test, nn_pred)

print('\nNeural Network Metrics:')
print(f'Accuracy: {nn_accuracy:.4f}')
print(f'F1 Score: {nn_f1:.4f}')
print(f'Precision: {nn_precision:.4f}')
print(f'Recall: {nn_recall:.4f}')
print('Confusion Matrix:')
print(nn_confusion_matrix)
```

```python
# Step 3: Display Evaluation Metrics
print('\nRandom Forest Metrics:')
print(f'Accuracy: {rf_accuracy:.4f}')
print(f'F1 Score: {rf_f1:.4f}')
print(f'Precision: {rf_precision:.4f}')
print(f'Recall: {rf_recall:.4f}')
print('Confusion Matrix:')
# Compute Confusion Matrix for Random Forest
rf_confusion_matrix = confusion_matrix(y_test, rf_pred)

print('\nRandom Forest Confusion Matrix:')
print(rf_confusion_matrix)
```

```python
# Function to generate and display a detailed confusion matrix
def plot_confusion_matrix(y_true, y_pred, classes, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.title(title)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
    # Confusion Matrix for Random Forest
plot_confusion_matrix(y_test, rf_pred, ['Benign', 'DDoS'], 'Random Forest Confusion Matrix')
```

```python
print('\nLogistic Regression Metrics:')
print(f'Accuracy: {lr_accuracy:.4f}')
print(f'F1 Score: {lr_f1:.4f}')
print(f'Precision: {lr_precision:.4f}')
print(f'Recall: {lr_recall:.4f}')
print('Confusion Matrix:')
# Compute Confusion Matrix for Logistic Regression
lr_confusion_matrix = confusion_matrix(y_test, lr_pred)

print('\nLogistic Regression Confusion Matrix:')
print(lr_confusion_matrix)
# Confusion Matrix for Logistic Regression
plot_confusion_matrix(y_test, lr_pred, ['Benign', 'DDoS'], 'Logistic Regression Confusion Matrix')
```

```python
print('\nNeural Network Metrics:')
print(f'Accuracy: {nn_accuracy:.4f}')
print(f'F1 Score: {nn_f1:.4f}')
print(f'Precision: {nn_precision:.4f}')
print(f'Recall: {nn_recall:.4f}')
print('Confusion Matrix:')
print(nn_confusion_matrix)
# Confusion Matrix for Neural Network
plot_confusion_matrix(y_test, nn_pred, ['Benign', 'DDoS'], 'Neural Network Confusion Matrix')
```

```python
# Random Forest
rf_proba = rf_model.predict_proba(X_test)

# Logistic Regression
lr_proba = lr_model.predict_proba(X_test)



# Neural Network
nn_proba = nn_model.predict_proba(X_test)
```

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc
# Convert categorical labels to binary
label_binarizer = LabelBinarizer()
y_test_binary = label_binarizer.fit_transform(y_test)

# Calculate ROC curve for Random Forest
rf_fpr, rf_tpr, _ = roc_curve(y_test_binary, rf_proba[:, 1])
rf_auc = auc(rf_fpr, rf_tpr)

# Calculate ROC curve for Logistic Regression
lr_fpr, lr_tpr, _ = roc_curve(y_test_binary, lr_proba[:, 1])
lr_auc = auc(lr_fpr, lr_tpr)

# Calculate ROC curve for Neural Network
nn_fpr, nn_tpr, _ = roc_curve(y_test_binary, nn_proba[:, 1])
nn_auc = auc(nn_fpr, nn_tpr)
```

```python
# Plot ROC curves for all models
plt.figure(figsize=(8, 6))
plt.plot(rf_fpr, rf_tpr, label=f'Random Forest (AUC = {rf_auc:.2f})')
plt.plot(lr_fpr, lr_tpr, label=f'Logistic Regression (AUC = {lr_auc:.2f})')
plt.plot(nn_fpr, nn_tpr, label=f'Neural Network (AUC = {nn_auc:.2f})')

# Plot ROC curve for random classifier (50% area)
plt.plot([0, 1], [0, 1], linestyle='--', color='black', label='Random Classifier (AUC = 0.50)')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid()
plt.show()
```
